```python
import pandas as pd
from sqlalchemy import create_engine


df = pd.read_csv(r'C:\Users\apurv\Documents\DataAnalystProjects\data\Sample - Superstore.csv', encoding='latin1')

df.columns = df.columns.str.replace(' ', '_').str.replace('-', '_')

df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'])

df['Year'] = df['Order_Date'].dt.year
df['Month_Num'] = df['Order_Date'].dt.month
df['Month_Name'] = df['Order_Date'].dt.month_name()
df['Delivery_Days'] = (df['Ship_Date'] - df['Order_Date']).dt.days

print("Data ready!")
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
```

```python
from sqlalchemy import create_engine
from urllib.parse import quote_plus

password = quote_plus('Apurav915@')

engine = create_engine(f'mysql+pymysql://root:{password}@localhost/superstore_db')

df.to_sql('sales', engine, if_exists='replace', index=False)

print("Data loaded into MySQL successfully!")
print("Table 'sales' created with", len(df), "rows")
```
