```python
def calculate_profit_margin(sales, profit):
    margin = (profit / sales) * 100
    return round(margin, 2)

margin1 = calculate_profit_margin(500, 75)
margin2 = calculate_profit_margin(1200, 96)
margin3 = calculate_profit_margin(300, -45)

print(f"Order 1 margin: {margin1}%")
print(f"Order 2 margin: {margin2}%")
print(f"Order 3 margin: {margin3}%")
```

```python
def classify_order(profit):
    if profit > 500:
        return "High Profit"
    elif profit > 0:
        return "Low Profit"
    elif profit == 0:
        return "Break Even"
    else:
        return "Loss"

# Test it
test_profits = [850, 120, 0, -200]
for p in test_profits:
    print(f"Profit ${p}: {classify_order(p)}")
```

```python
import pandas as pd

df = pd.read_csv(r'C:\Users\apurv\Documents\DataAnalystProjects\data\Sample - Superstore.csv', encoding='latin1')
print("Data loaded! Shape:", df.shape)
```

```python
total_rows = df.shape[0]
total_columns = df.shape[1]
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
avg_order_value = df['Sales'].mean()

print(f"Total Orders: {total_rows:,}")
print(f"Total Revenue: ${total_sales:,.2f}")
print(f"Total Profit: ${total_profit:,.2f}")
print(f"Average Order Value: ${avg_order_value:,.2f}")
```

```python
regions = df['Region'].unique().tolist()
categories = df['Category'].unique().tolist()

print("Regions in dataset:", regions)
print("Categories in dataset:", categories)
print("Number of unique customers:", df['Customer Name'].nunique())
```

```python
def analyze_column(column_name):
    col = df[column_name]
    print(f"Analysis of '{column_name}':")
    print(f"  Total:   ${col.sum():,.2f}")
    print(f"  Average: ${col.mean():,.2f}")
    print(f"  Highest: ${col.max():,.2f}")
    print(f"  Lowest:  ${col.min():,.2f}")
    print()

# Run on Sales and Profit
analyze_column('Sales')
analyze_column('Profit')
```

```python
print("Sales by Region:")
for region in df['Region'].unique():
    region_data = df[df['Region'] == region]
    total = region_data['Sales'].sum()
    print(f"  {region}: ${total:,.2f}")
```

```python
print("Category Profitability Report:")
for category in df['Category'].unique():
    cat_data = df[df['Category'] == category]
    profit = cat_data['Profit'].sum()
    sales = cat_data['Sales'].sum()
    margin = (profit / sales) * 100

    if margin > 15:
        status = "Excellent"
    elif margin > 10:
        status = "Good"
    elif margin > 0:
        status = "Low"
    else:
        status = "Loss Making"

    print(f"  {category}: {round(margin,1)}% margin → {status}")
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv(r'C:\Users\apurv\Documents\DataAnalystProjects\data\Sample - Superstore.csv', encoding='latin1')

print("Data loaded! Shape:", df.shape)
df.head()
```

```python
sales_col = df['Sales']

print(type(sales_col))  
print(sales_col.head(5))
```

```python

subset = df[['Customer Name', 'Region', 'Sales', 'Profit']]

print("Shape of subset:", subset.shape)
subset.head(10)
```

```python
number_cols = df.select_dtypes(include='number')
print("Number columns:", number_cols.columns.tolist())

text_cols = df.select_dtypes(include='object')
print("Text columns:", text_cols.columns.tolist())
```

```python
east_orders = df[df['Region'] == 'East']

print("Total orders in East:", len(east_orders))
print("East region sales: $", round(east_orders['Sales'].sum(), 2))
```

```python
high_value = df[df['Sales'] > 1000]

print("High value orders:", len(high_value))
high_value[['Customer Name', 'Category', 'Sales', 'Profit']].head(10)
```

```python

losses = df[df['Profit'] < 0]

print("Loss making orders:", len(losses))
print("That is", round(len(losses)/len(df)*100, 1), "% of all orders")
print("Total money lost: $", round(losses['Profit'].sum(), 2))
```

```python

good_orders = df[(df['Sales'] > 500) & (df['Profit'] > 0)]
print("High sales + profitable orders:", len(good_orders))

tech_furniture = df[(df['Category'] == 'Furniture') | (df['Category'] == 'Technology')]
print("Furniture or Tech orders:", len(tech_furniture))
```

```python
phone_orders = df[df['Product Name'].str.contains('Phone', case=False)]

print("Phone related orders:", len(phone_orders))
print("Total phone sales: $", round(phone_orders['Sales'].sum(), 2))
phone_orders[['Product Name', 'Sales', 'Profit']].head(5)
```

```python

top_sales = df.sort_values('Sales', ascending=False)

print("Top 10 highest sales orders:")
top_sales[['Customer Name', 'Product Name', 'Sales', 'Profit']].head(10)
```

```python

sorted_df = df.sort_values(['Region', 'Sales'], ascending=[True, False])

sorted_df[['Region', 'Customer Name', 'Sales']].head(10)
```

```python
print("TOP 5 MOST PROFITABLE ORDERS:")
top5 = df.nlargest(5, 'Profit')[['Customer Name', 'Product Name', 'Sales', 'Profit']]
print(top5.to_string())

print()


print("TOP 5 BIGGEST LOSS ORDERS:")
bottom5 = df.nsmallest(5, 'Profit')[['Customer Name', 'Product Name', 'Sales', 'Profit']]
print(bottom5.to_string())
```

```python

region_sales = df.groupby('Region')['Sales'].sum()

print("Total Sales by Region:")
print(region_sales.sort_values(ascending=False))
```

```python
category_stats = df.groupby('Category').agg(
    Total_Sales   = ('Sales', 'sum'),
    Total_Profit  = ('Profit', 'sum'),
    Total_Orders  = ('Order ID', 'count'),
    Avg_Sale      = ('Sales', 'mean')
).round(2)

print("Category Performance:")
print(category_stats)
```

```python

region_category = df.groupby(['Region', 'Category'])['Sales'].sum().reset_index()
region_category = region_category.sort_values('Sales', ascending=False)

print("Sales by Region and Category:")
print(region_category.head(12).to_string())
```

```python
summary = df.groupby('Category').agg(
    Sales  = ('Sales', 'sum'),
    Profit = ('Profit', 'sum')
).round(2)


summary['Profit_Margin_%'] = ((summary['Profit'] / summary['Sales']) * 100).round(1)

print("Category Summary with Profit Margin:")
print(summary)
```

```python

df['Profit_Margin'] = ((df['Profit'] / df['Sales']) * 100).round(2)


df['Order_Size'] = pd.cut(
    df['Sales'],
    bins=[0, 100, 500, 1000, 100000],
    labels=['Small', 'Medium', 'Large', 'Very Large']
)


df['Year'] = pd.to_datetime(df['Order Date']).dt.year
df['Month'] = pd.to_datetime(df['Order Date']).dt.month_name()

df[['Sales', 'Profit', 'Profit_Margin', 'Order_Size', 'Year', 'Month']].head(10)
```

```python

print("Orders per Region:")
print(df['Region'].value_counts())

print()

print("Orders per Category:")
print(df['Category'].value_counts())

print()

print("Orders per Segment:")
print(df['Segment'].value_counts())
```

```python
print("=" * 50)
print("   SUPERSTORE SALES ANALYSIS REPORT")
print("=" * 50)

print("\n1. OVERALL PERFORMANCE")
print(f"   Total Orders:  {len(df):,}")
print(f"   Total Revenue: ${df['Sales'].sum():,.2f}")
print(f"   Total Profit:  ${df['Profit'].sum():,.2f}")
print(f"   Overall Margin: {round(df['Profit'].sum()/df['Sales'].sum()*100, 1)}%")


print("\n2. BEST PERFORMING REGION")
best = df.groupby('Region')['Sales'].sum().idxmax()
best_sales = df.groupby('Region')['Sales'].sum().max()
print(f"   {best} with ${best_sales:,.2f} in sales")


print("\n3. MOST PROFITABLE CATEGORY")
best_cat = df.groupby('Category')['Profit'].sum().idxmax()
best_profit = df.groupby('Category')['Profit'].sum().max()
print(f"   {best_cat} with ${best_profit:,.2f} in profit")


losses = df[df['Profit'] < 0]
print("\n4. LOSS ANALYSIS")
print(f"   Loss-making orders: {len(losses):,} ({round(len(losses)/len(df)*100,1)}%)")
print(f"   Total losses: ${losses['Profit'].sum():,.2f}")


print("\n5. TOP CUSTOMER")
top_cust = df.groupby('Customer Name')['Sales'].sum().idxmax()
top_cust_sales = df.groupby('Customer Name')['Sales'].sum().max()
print(f"   {top_cust} with ${top_cust_sales:,.2f} in purchases")

print("\n" + "=" * 50)
print("   END OF REPORT")
print("=" * 50)
```
