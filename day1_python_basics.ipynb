```python
def calculate_profit_margin(sales, profit):
    margin = (profit / sales) * 100
    return round(margin, 2)

margin1 = calculate_profit_margin(500, 75)
margin2 = calculate_profit_margin(1200, 96)
margin3 = calculate_profit_margin(300, -45)

print(f"Order 1 margin: {margin1}%")
print(f"Order 2 margin: {margin2}%")
print(f"Order 3 margin: {margin3}%")
```

```python
def classify_order(profit):
    if profit > 500:
        return "High Profit"
    elif profit > 0:
        return "Low Profit"
    elif profit == 0:
        return "Break Even"
    else:
        return "Loss"

# Test it
test_profits = [850, 120, 0, -200]
for p in test_profits:
    print(f"Profit ${p}: {classify_order(p)}")
```

```python
import pandas as pd

df = pd.read_csv(r'C:\Users\apurv\Documents\DataAnalystProjects\data\Sample - Superstore.csv', encoding='latin1')
print("Data loaded! Shape:", df.shape)
```

```python
total_rows = df.shape[0]
total_columns = df.shape[1]
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
avg_order_value = df['Sales'].mean()

print(f"Total Orders: {total_rows:,}")
print(f"Total Revenue: ${total_sales:,.2f}")
print(f"Total Profit: ${total_profit:,.2f}")
print(f"Average Order Value: ${avg_order_value:,.2f}")
```

```python
regions = df['Region'].unique().tolist()
categories = df['Category'].unique().tolist()

print("Regions in dataset:", regions)
print("Categories in dataset:", categories)
print("Number of unique customers:", df['Customer Name'].nunique())
```

```python
def analyze_column(column_name):
    col = df[column_name]
    print(f"Analysis of '{column_name}':")
    print(f"  Total:   ${col.sum():,.2f}")
    print(f"  Average: ${col.mean():,.2f}")
    print(f"  Highest: ${col.max():,.2f}")
    print(f"  Lowest:  ${col.min():,.2f}")
    print()

# Run on Sales and Profit
analyze_column('Sales')
analyze_column('Profit')
```

```python
print("Sales by Region:")
for region in df['Region'].unique():
    region_data = df[df['Region'] == region]
    total = region_data['Sales'].sum()
    print(f"  {region}: ${total:,.2f}")
```

```python
print("Category Profitability Report:")
for category in df['Category'].unique():
    cat_data = df[df['Category'] == category]
    profit = cat_data['Profit'].sum()
    sales = cat_data['Sales'].sum()
    margin = (profit / sales) * 100

    if margin > 15:
        status = "Excellent"
    elif margin > 10:
        status = "Good"
    elif margin > 0:
        status = "Low"
    else:
        status = "Loss Making"

    print(f"  {category}: {round(margin,1)}% margin → {status}")
```
